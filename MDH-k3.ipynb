```python
#Fits data to  y = a*(1+x/(d+x))+b*x/(d+x)
```

```python
import numpy as np
import scipy as sp
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
#Defining the equation.
def ez_reaction (x, a, b, d):
  y = a*(1+x/(d+x))+b*x/(d+x)
  return y
```

```python
#Experimental data
realx = np.array([0, 5, 15, 30])
realk3 = [0.68, 5.85, 10.91, 13.9]
K = 0.68, 1, 1 #initial guess
#fitting with curve fit
params, pcov  = curve_fit(ez_reaction, realx, realk3, p0 = K, method='lm')
print(params)
perr = np.sqrt(np.diag(pcov))
print(perr) #calculate SD
```

```python
#ploting the data and the fit
x = np.arange(0, 31, 1)
y = ez_reaction (x, *params)
plt.plot(x, y)
plt.plot(realx, realk3, 'o')
plt.xlabel('OAD (units)')
plt.ylabel('k3 (1/s)')
```
